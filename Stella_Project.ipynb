##### Histogram of Star Ratings

```python
# histogram for the distribution of stars
plt.figure(figsize=(10, 6))  # sets the figure size

sns.histplot(kindle2_df['stars'], 
             bins=25,               # size of the bins
             color='teal',         # bar colour
             edgecolor='black')  # Adds borders to the bars 

# Adds labels and title
plt.title('Distribution of Stars', fontsize=14)
plt.xlabel('Stars', fontsize=12)
plt.ylabel('Number of Books', fontsize=12)

# Show the plot
plt.show()
```

##### A Histogram showing the Distribution of Book Prices

```python
# histogram for the distribution of price
plt.figure(figsize=(12, 6))     # sets the figure size
plt.hist(kindle2_df['price'], 
         bins=25,               # size of bins
         color='plum',          # bar colour 
         edgecolor='black')     # adds borders to the bars

# Adds labels and title
plt.xlabel('Price', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.title('Distribution of Book Prices', fontsize=14)

# Show the plot
plt.show()
```

#### Calculate the average book price per category

```python
# Calculate the average price per category
# Group by 'category_name' and calculate the average of 'price'
avg_price_per_category = kindle2_df.groupby('category_name')['price'].mean().reset_index()     # reset the index and return back as a column

avg_price_per_category
```

```python
# Create a bar chart using Seaborn of average price per category
plt.figure(figsize=(12, 6))
sns.barplot(x='category_name', y='price', data=avg_price_per_category, hue='category_name', palette='pastel', legend=False)

# Add labels and title
plt.xlabel('Category Name', fontsize=14)
plt.ylabel('Average Price', fontsize=14)
plt.title('Average Price per Category', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=10)

# Show the plot
plt.show()
```

#### How does the book price relate to the number of stars?

```python
# scatter plot of price vs stars using Seaborn
plt.figure(figsize=(12, 6))
sns.scatterplot(x='stars', y='price', data=kindle2_df, color='purple')

# Adds labels and title
plt.xlabel('Stars', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.title('Scatter Plot of Price vs Stars', fontsize=14)

# Show the plot
plt.show()
```

#### Scatter plot showing stars vs number of reviews

#### A bar chart showing all the James Patterson books by category

```python
# Set figure size
plt.figure(figsize=(12, 6))  # Width = 12 inches, Height = 6 inches

# Create count plot and Use a built-in Seaborn color palette
sns.countplot(x='category_name', data=JPatterson, hue='category_name',palette="magma", legend=False)

# Add labels
plt.xlabel("Category Name", fontsize=12)  # X-axis label
plt.ylabel("Count", fontsize=12)          # Y-axis label
plt.title("James Patterson books by category", fontsize=14)  # Title

# Rotate x-axis labels for better readability
plt.xticks(rotation=90)  # Rotate labels by 90 degrees

# Show plot
plt.show()
```

#### A stacked bar chart showing the number of stars by category for James Patterson books

```python
# Count the number of books in each category for each star rating
count_stars = JPatterson.groupby(['category_name', 'stars']).size().unstack(fill_value=0)

# Create a stacked bar chart
count_stars.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='tab10')

# Add labels and title
plt.xlabel('Category', fontsize=14)
plt.ylabel('Number of Books', fontsize=14)
plt.title('Stacked Bar Chart - James Patterson books - Number of Stars per Category', fontsize=14)

# Show the plot
plt.show()
```

#### The average James Patterson book price per category and star rating
stacked bar chart

```python
# Group by category_name and stars, then calculate the average price
agg_df = JPatterson.groupby(['category_name', 'stars']).agg({'price': 'mean'}).reset_index()    # using Aggregate to calculate the average (mean) price 

# Create a bar chart of the average price per category and star rating
fig = px.bar(agg_df, x='category_name', y='price', color='stars', 
             title='James Patterson - Average Book Price per Category and Stars',
             labels={'price': 'Average Price', 'category_name': 'Category', 'stars': 'Stars'})

# Show the figure
fig.show()
```

##### Pie Chart showing the percentage of Kindle Unlimited vs None Kindle Unlimited books

```python
# Plotting the pie chart
plt.figure(figsize=(5, 5))  # Set the figure size

kindle2_df['isKindleUnlimited'].value_counts().plot(kind='pie', 
                                                     autopct='%.2f%%', 
                                                     labels=['Not KU', 'Is KU'],
                                                     startangle=90,   # Start the chart at a specific angle
                                                     colors=['#ff9999', '#66b3ff'],  
                                                     wedgeprops={'edgecolor': 'black', 'linewidth': 1, 'linestyle': 'solid'},  # Add border to wedges
                                                     explode=(0.1, 0))  # "explode" the second slice for emphasis

# Title and labels 
plt.title('Percentage of Books that are Kindle Unlimited', fontsize=16, fontweight='bold', color='darkblue')

# Display the chart
plt.axis('equal')  # Ensures the pie chart is a perfect circle
plt.show()
```

##### Bar Chart showing how many books per category

```python
book_categories = kindle2_df['category_name'].value_counts().reset_index()

# bar chart showing books per category
plt.figure(figsize=(18, 6))     # sets the figure size

sns.barplot(data = book_categories,  
         x = "category_name",
         y = 'count',
         hue='category_name', palette="dark",           # bar colour 
         edgecolor='black')     # adds borders to the bars

# Adds labels and title
plt.xlabel('Book Category', fontsize=12)
plt.ylabel('Book Count', fontsize=12)
plt.title('Books by Category', fontsize=14)
plt.xticks(rotation=90)

# Show the plot
plt.show()
```

#### How many Best Seller books are there on the list?

* There are 2,236 Best Sellers on the list, out of 132,000 books.

```python
# how many best selling books in each category

bestseller_list = kindle2_df[kindle2_df['isBestSeller'] == True]  # show which books are Best Sellers

bestseller_counts = bestseller_list['category_name'].value_counts().reset_index() # Count occurrences of best sellers per category
print(bestseller_counts.sum())     # sums values to give total no


# Create a bar chart using Seaborn of Best Sellers per category
plt.figure(figsize=(12, 6))     # set figure size
sns.barplot(x='category_name', 
            y='count', 
            data=bestseller_counts, 
            hue='category_name', palette='dark', 
            edgecolor='black')

# Add labels and title
plt.xlabel('Category Name', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.title('Best Sellers per Category', fontsize=16)

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=10)

# Show the plot
plt.show()
```
